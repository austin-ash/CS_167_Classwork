## Normalization Code

```
titanic = pandas.read_csv('/content/drive/MyDrive/CS167/datasets/titanic.csv')
titanic.head()
```

```
#convert sex to numeric
titanic['sex'] = titanic['sex'].replace(to_replace='female', value=1)
titanic['sex'] = titanic['sex'].replace(to_replace='male', value=0)
titanic.head()
```

```
#gender
#calculate mean and std
s_mean = titanic['sex'].mean()
s_std = titanic['sex'].std()

#replace column with each entry's z-score
titanic['sex'] = (titanic["sex"] - s_mean)/s_std
titanic.head()
```

```
#age
#calculate mean and std
a_mean = titanic['age'].mean()
a_std = titanic['age'].std()

#replace column with each entry's z-score
titanic['age'] = (titanic["age"] - a_mean)/a_std
titanic.head()
```

## kNN Functions:

```
def knn(specimen, data, k):
  data['distance_to_new'] = numpy.sqrt(
    (specimen['petal length']-data['petal length'])**2
    +(specimen['sepal length']-data['sepal length'])**2
    +(specimen['petal width']-data['petal width'])**2
    +(specimen['sepal width']-data['sepal width'])**2)
  
  sorted_data = data.sort_values(['distance_to_new']) 
  return sorted_data.iloc[0:5]['species'].mode()[0]
```

```
def w_knn(specimen, data, k):
  #calculate the distance
  data['distance_to_new'] = numpy.sqrt(
    (specimen['petal length']-data['petal length'])**2
    +(specimen['sepal length']-data['sepal length'])**2
    +(specimen['petal width']-data['petal width'])**2
    +(specimen['sepal width']-data['sepal width'])**2)
  
  #calculate the weights (remember, the weights are 1/d^2)
  data['weight'] = 1.0/(data['distance_to_new']**2)

  #find the closest k
  sorted_data = data.sort_values(['distance_to_new']) 

  print(sorted_data.iloc[0:k]) #take a look at the results

  #sum the weights and return the class of the larger sum. 
  print(sorted_data.iloc[0:k].groupby(['species'])['weight'].sum())
  return sorted_data.iloc[0:k].groupby(['species'])['weight'].sum().idxmax()
```

```
iris = pandas.read_csv('/content/drive/MyDrive/CS167/datasets/irisData.csv')
iris.head()
```

```
# Petal length
#calculate mean and std
plength_mean = iris['petal length'].mean()
plength_std = iris['petal length'].std()

#replace column with each entry's z-score
iris['petal length'] = (iris["petal length"] - plength_mean)/plength_std
iris.head()
```

```
# sepal length
#calculate mean and std
slength_mean = iris['sepal length'].mean()
slength_std = iris['sepal length'].std()

#replace column with each entry's z-score
iris['sepal length'] = (iris["sepal length"] - slength_mean)/slength_std
iris.head()
```

```
# Petal width
#calculate mean and std
pwidth_mean = iris['petal width'].mean()
pwidth_std = iris['petal width'].std()

#replace column with each entry's z-score
iris['petal width'] = (iris["petal width"] - pwidth_mean)/pwidth_std
iris.head()
```

```
# Sepal width
#calculate mean and std
swidth_mean = iris['sepal width'].mean()
swidth_std = iris['sepal width'].std()

#replace column with each entry's z-score
iris['sepal width'] = (iris["sepal width"] - swidth_mean)/swidth_std
iris.head()
```

```
# normalized dictionary of values
norm_specimen = {}
norm_specimen['petal length'] = (1.5 - plength_mean)/plength_std
norm_specimen['sepal length'] = (0.5 - slength_mean)/slength_std
norm_specimen['petal width'] = (2.0 - pwidth_mean)/pwidth_std
norm_specimen['sepal width'] = (1.7 - swidth_mean)/swidth_std
# 5-NN 
knn(norm_specimen,iris,5)
```

```
# Original dictionary of values
specimen = {}
specimen['petal length'] = 5.1
specimen['sepal length'] = 7.2
specimen['petal width'] = 1.5
specimen['sepal width'] = 2.5
# 5-NN 
knn(specimen,iris,5)
```
